##  **1.-Carga y visualizacion.**

```python
# Importar librerías necesarias
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans, AgglomerativeClustering
from sklearn.metrics import silhouette_score, davies_bouldin_score, calinski_harabasz_score
from sklearn.metrics import davies_bouldin_score
from datetime import datetime
from scipy.cluster.hierarchy import dendrogram, linkage
```

```python
# Cargar datos
file_path = "Retail_Invoices.xlsx"
df = pd.read_excel(file_path)
```

```python
#Visualizar  DF
df.head()
```

```python
# Visualizamos los tipos de datos
print(df.info())
```

```python
#Buscamos valores nulos
df.isnull().sum()
```

## **2.-Preprocesamiento**

```python
# Eliminar registros con cantidades cero o negativas
df = df[df['Quantity'] > 0]
```

```python
# Verificamos valores duplicados por boleta-producto
df.drop_duplicates(inplace=True)
```

```python
# RFM por cliente.
snapshot_date = df['InvoiceDate'].max() + pd.Timedelta(days=1)

rfm = df.groupby('CustomerID').agg({
    'InvoiceDate': lambda x: (snapshot_date - x.max()).days,  # Recency
    'InvoiceNo': 'nunique',                                    # Frequency
    'price_total': 'sum'                                       # Monetary (con devoluciones)
}).reset_index()

rfm.columns = ['CustomerID', 'Recency', 'Frequency', 'Monetary']
```

```python
rfm.head()
```

```python
# Análisis exploratorio de RFM
sns.pairplot(rfm[['Recency', 'Frequency', 'Monetary']])
plt.suptitle("Distribución de variables RFM", y=1.02)
plt.show()
```

1. Diagonales (histogramas):  Los datos están muy sesgados a la derecha, con muchos valores bajos y pocos valores extremos muy altos. Eso es normal en comportamiento de clientes (la regla del 80/20 o de Pareto).
- Recency: La mayoría de los clientes compraron hace pocos días (valores bajos), y solo unos pocos tienen Recency muy alta (no han comprado en mucho tiempo).
- Frequency: La mayoría de los clientes compraron pocas veces. Algunos (muy pocos) hicieron muchas compras.
- Monetary: Casi todos gastaron poco, pero hay algunos clientes que gastaron muchísimo (outliers que llegan a más de 250,000).

2. Recency vs Frequency
Se ve una nube de puntos agrupada en la esquina inferior izquierda. Clientes con recency baja (compraron recientemente) tienden a tener frecuencia más alta.Hay una tendencia débil: quienes compran más seguido, también lo hacen más recientemente. Pero no es una regla estricta.

3. Recency vs Monetary: Algunos clientes tienen recency baja y gastaron mucho, pero hay bastante dispersión.Clientes que compraron hace poco no siempre son los que más gastan. No hay una correlación clara.

4. Frequency vs Monetary: Aquí sí se nota una tendencia más clara: a mayor frecuencia, en general mayor gasto.Pero hay excepciones (clientes que compraron pocas veces pero gastaron mucho).

```python
# Excluir CustomerID antes de calcular la correlación
rfm_corr = rfm.drop(columns=['CustomerID']).corr()
# Visualizar heatmap sin CustomerID
sns.heatmap(rfm_corr, annot=True, cmap='coolwarm')
plt.title("Correlación entre variables RFM")
plt.show()
```

**Conclusión del análisis de correlación:**
- Frequency y Monetary están correlacionadas de manera mediocre.
- Recency es una dimensión más independiente, lo cual es ideal en modelos de clustering, porque aporta una vista diferente (tiempo).
- No hay multicolinealidad fuerte.

```python
# Escalado de variables
scaler = StandardScaler()
rfm_scaled = scaler.fit_transform(rfm[['Recency', 'Frequency', 'Monetary']])
rfm_scaled
```

##  **3.-Modelos Clustering: KMeans y Agglomerative**

### **3.1- Busqueda del cluster optimo**

```python
# Determinar número óptimo de clusters con KMeans (codo , silhouette, Davis Bouldin)
inertia = []
sil_scores = []
db_scores = []
k_range = range(2, 10)

for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(rfm_scaled)
    inertia.append(kmeans.inertia_)
    sil_scores.append(silhouette_score(rfm_scaled, kmeans.labels_))
    db_scores.append(davies_bouldin_score(rfm_scaled,kmeans.labels_))


# Gráfico del codo
plt.plot(k_range, inertia, marker='o')
plt.title('Método del Codo - Inercia')
plt.xlabel('Número de Clusters')
plt.ylabel('Inercia')
plt.show()

# Gráfico de Silhouette
plt.plot(k_range, sil_scores, marker='o')
plt.title('Puntaje Silhouette')
plt.xlabel('Número de Clusters')
plt.ylabel('Silhouette Score')
plt.show()

# Gráfico de Silhouette
plt.plot(k_range, db_scores, marker='o')
plt.title('Davies- Bouldin')
plt.xlabel('Número de Clusters')
plt.ylabel('Bouldin Score')
plt.show()
```

**Conclusion:**

Parece que el rango optimo se encuentra entre 4 y 5, aunque haria falta hacer mas pruebas para concluir.

```python
# Determinar número óptimo de clusters con AGG (Calinski Score , silhouette, Davis Bouldin)
k_range = range(2, 10)

sil_scores = []
db_scores  = []
ch_scores  = []

for k in k_range:
    # Modelo jerárquico
    agg = AgglomerativeClustering(n_clusters=k, linkage='ward')
    agg.fit_predict(rfm_scaled)
    sil_scores.append(silhouette_score(rfm_scaled, agg.labels_))
    db_scores.append(davies_bouldin_score(rfm_scaled, agg.labels_))
    ch_scores.append(calinski_harabasz_score(rfm_scaled,  agg.labels_))

# Gráfico
plt.plot(k_range, ch_scores, marker='o')
plt.title('Puntaje calinski- Harabaz')
plt.xlabel('Número de Clusters')
plt.ylabel('Calinski Score')
plt.show()

# Gráfico de Silhouette
plt.plot(k_range, sil_scores, marker='o')
plt.title('Puntaje Silhouette')
plt.xlabel('Número de Clusters')
plt.ylabel('Silhouette Score')
plt.show()

# Gráfico de Silhouette
plt.plot(k_range, db_scores, marker='o')
plt.title('Davies- Bouldin')
plt.xlabel('Número de Clusters')
plt.ylabel('Bouldin Score')
plt.show()
```

```python
Z = linkage(rfm_scaled, method='ward')
dendrogram(Z)
plt.title("Dendrograma")
plt.xlabel("Índice de muestra")
plt.ylabel("Distancia")
plt.show()
```

**Conclusion:**
El numero de cluster parece en el rango 2 a 4. Esto es una suposicion preliminar, hacen falta mas pruebas para concluir satisfactoriamente.

### **3.2- Creacion del Modelo**

Después de realizar múltiples pruebas con un rango de K(2, 4) en ambos algoritmos, y comparar los resultados mediante diferentes gráficos de las variables, se llegó a la conclusión satisfactoria de que K= 3 son los óptimos. Ambos algoritmos convergen en esta solución y ofrecen una interpretación clara y consistente. En las conclusiones finales y notas se explica este proceso con mayor detalle.

```python
#  Aplicar KMeans con n=3 
kmeans = KMeans(n_clusters=3, random_state=42)
rfm['Cluster_KMeans'] = kmeans.fit_predict(rfm_scaled)

# Aplicar clustering jerárquico
agg = AgglomerativeClustering(n_clusters=3)
rfm['Cluster_Agg'] = agg.fit_predict(rfm_scaled)
```

```python
# Visualización de clusters con K-Means y Agglomerative: Usando Monetary y Recency
sns.scatterplot(data=rfm, x='Recency', y='Monetary', hue='Cluster_KMeans', palette='tab10')
plt.title('Clusters por KMeans')
plt.show()

sns.scatterplot(data=rfm, x='Recency', y='Monetary', hue='Cluster_Agg', palette='Set2')
plt.title('Clusters por Agglomerative Clustering')
plt.show()
```

```python
#Visualización de clusters con K-Means y Agglomerative: Usando Monetary y Frequency
sns.scatterplot(data=rfm, x='Frequency', y='Monetary', hue='Cluster_KMeans', palette='tab10')
plt.title('Clusters por KMeans')
plt.show()

sns.scatterplot(data=rfm, x='Frequency', y='Monetary', hue='Cluster_Agg', palette='Set2')
plt.title('Clusters por Agglomerative Clustering')
plt.show()
```

```python
# 11. Evaluar clustering
score_kmeans_sil = silhouette_score(rfm_scaled, rfm['Cluster_KMeans'])
score_agg_sil = silhouette_score(rfm_scaled, rfm['Cluster_Agg'])

score_kmeans_boul = davies_bouldin_score(rfm_scaled, rfm['Cluster_KMeans'])
score_agg_boul = davies_bouldin_score(rfm_scaled, rfm['Cluster_Agg'])

print(f"Silhouette Score - KMeans: {score_kmeans_sil:.2f}")
print(f"Silhouette Score - Agglomerative: {score_agg_sil:.2f}")
print(f"Davis Bouldin - KMeans: {score_kmeans_boul:.2f}")
print(f"Davis Bouldin - Agglomerative: {score_agg_boul:.2f}")
```

```python
# 12. Interpretación de segmentos (nombres ejemplo)
segmentos = rfm.groupby('Cluster_Agg').agg({
    'Recency': 'mean',
    'Frequency': 'mean',
    'Monetary': 'mean',
    'CustomerID': 'count'
}).rename(columns={'CustomerID': 'N_Clientes'})

print("\nResumen de segmentos:")
segmentos  
```

**Cluster 1 – Cliente Leal**
Clientes con compras muy recientes, frecuencia altísima y valor monetario muy elevado.

**Cluster 0 – Clientes Recurrente**
Compras relativamente recientes y moderadas, clientes fieles pero con margen de crecimiento.

**Cluster 2 – Cliente ocacional**
Hace mucho que no compran (alta recencia), pocas compras y bajo gasto. Requieren campañas de reactivación.

#### **Conclusión Final – Segmentación RFM**
Después de realizar múltiples pruebas con distintos valores de   k (de 2 a 4 clusters) utilizando tanto KMeans como Agglomerative Clustering, y evaluando el rendimiento mediante las métricas Silhouette Score, Davies–Bouldin Index y análisis visual de las variables RFM por cluster, se llegó a una conclusión sólida:

El número óptimo de clusters es 3, ya que:
- Ambos algoritmos convergen en segmentaciones consistentes con buena separación.
- Las métricas cuantitativas (Silhouette ≈ 0.59 y DBI ≈ 0.70) reflejan buena calidad de agrupamiento.
- El análisis gráfico de las variables Recency, Frequency y Monetary por grupo muestra diferencias claras y significativas entre los segmentos.

Esta convergencia entre métodos valida la estabilidad del modelo y proporciona una base sólida para la estrategia de marketing y personalización basada en segmentos de clientes.
