Importamos las librerias Necesarias

```python
import pandas as pd
from pycaret.datasets import get_data
```

Cargamos el Dataset.

```python
dataset = get_data('credit')
```

En este caso elegimos una muestra del 95% de los datos del Datadaset

```python
data = dataset.sample(frac=.95, random_state=0)
data
```

```python
data.shape
```

El dataset Restante lo guardamos en otro DF, nos servirá como otro subconjunto para evaluar el modelo

```python
data_unseen = dataset.drop(data.index)
data_unseen
```

```python
#Reseteamos los indices
data.reset_index(inplace= True, drop=True)
data_unseen.reset_index(inplace=True, drop= True)
```

```python
#importamos la libreria para utilizar Pycaret
from pycaret.classification import *
```

```python
# Elegimos los parametros del Setup, intoducimos el dataset = data, y la variable objetivo Target= default, fix_imbalance= true para aplicar Smote para desblanceo de  clases
model_setup = setup(data= data, target="default", session_id=123 , fix_imbalance=True)
```

<p><strong data-start="119" data-end="154">Conclusi&oacute;n del setup en PyCaret</strong></p>
<p data-start="156" data-end="498">El experimento fue configurado para un problema de clasificaci&oacute;n binaria con el objetivo de predecir la variable <strong data-start="269" data-end="282">"default"</strong> en un dataset original de 22,800 registros y 24 caracter&iacute;sticas. Tras aplicar preprocesamiento, la muestra aument&oacute; a 31,721 registros, distribuidos en un conjunto de entrenamiento de 24,880 y uno de prueba de 6,841.</p>
<p data-start="500" data-end="798">Se usaron 23 variables num&eacute;ricas, con imputaci&oacute;n simple basada en la media para valores faltantes num&eacute;ricos y la moda para categ&oacute;ricos. Para corregir el desequilibrio de clases se aplic&oacute; la t&eacute;cnica <strong data-start="698" data-end="707">SMOTE</strong>. La validaci&oacute;n cruzada se realiz&oacute; con un generador de pliegues estratificados de 10 folds.</p>
<p data-start="800" data-end="917">El setup se ejecut&oacute; sin GPU y sin registro de experimentos externos, utilizando todos los n&uacute;cleos de CPU disponibles.</p>

## Lightgbm

```python
#Optimizamos el primero modelo.
lg = create_model("lightgbm")
```

```python
#Optimizamos el mismo modelo pero tratando de mejorar el recall y f1
lg2 = create_model("lightgbm", scale_pos_weight=3.5231813041839426)
```

```python
print(lg2)
```

```python
# Buscamos los mejores hiperparametros para el modelo original
tuned_lg = tune_model(lg)
```

```python
#Buscamos los mejores hiperparametros para el modelo que busca balancear mas ambas clases
tuned_lg2 = tune_model(lg2)
```

```python
#evaluacion del modelo original
evaluate_model(tuned_lg)
```

```python
#evaluacion del modelo con scale_pos_weight
evaluate_model (tuned_lg2)
```

<p><strong>Modelo original</strong></p>
<p>Mejor desempe&ntilde;o general: accuracy (82.2%), AUC (0.7789), F1 Score (0.4895), MCC (0.4119).</p>
<p>Pero su recall promedio es bajo (38.8%), lo que indica que pierde muchos casos positivos (clase 1).</p>
<p><strong>Modelo con scale_pos_weight</strong></p>
<p>Sacrifica algo de precisi&oacute;n global (accuracy baja a 75.9%), pero mejora notablemente el recall (57.0%), lo que significa que detecta m&aacute;s casos positivos.</p>
<p>Tiene un F1 Score ligeramente superior (0.5114), lo que refleja mejor balance entre precisi&oacute;n y recall.</p>

<h3 data-start="984" data-end="1005"><strong data-start="991" data-end="1005">Conclusi&oacute;n</strong></h3>
<ul data-start="1006" data-end="1276">
<li data-start="1006" data-end="1161">
<p data-start="1008" data-end="1161">Si es m&aacute;s importante detectar los casos positivos (clase 1), el modelo con scale_pos_weight es preferible por su mayor recall y mejor F1 Score.</p>
</li>
<li data-start="1162" data-end="1276">
<p data-start="1164" data-end="1276">Si se prioriza el desempe&ntilde;o global y el balance entre clases no es un problema, el modelo original es mejor.</p>

## Random Forest

```python
#Creamos el modelo Random Forest, aplicamos class weight tratando de balancear las clases
rf = create_model("rf", class_weight='balanced')
```

```python
#Observamos los hiperparametros que tiene originalmente
print (rf)
```

```python
#buscamos los hiperparametros optimos , optimizamos f1 para darle mas peso a la clase 1
tuned_rf = tune_model(rf, optimize='F1')
```

```python
#Hiperparametros optimos
print(tuned_rf)
```

```python
#Metricas del modelo Random Forest
evaluate_model(tuned_rf)
```

### Comparar Modelos

```python
#Metricas generadas con las predicciones del conjunto de prueba que nos separo pycaret (30% del dataset)
# Predicción con el primer modelo (Lightgmb)
results1_1 = predict_model(tuned_lg)
metrics1_1 = pull()  # <-- inmediatamente después

# Predicción con el segundo modelo (Random Forest) 
results2_1 = predict_model(tuned_rf)
metrics2_1 = pull()  # <-- inmediatamente después
```

```python
#Impresion de metricas
print(metrics1_1)
print(metrics2_1)
```

```python
#Metricas generadas con las predicciones del conjunto que separamos inicialmente.
# Predicción con el primer modelo
results1_2 = predict_model(tuned_lg, data=data_unseen)
metrics1_2 = pull()  # <-- inmediatamente después

# Predicción con el segundo modelo
results2_2 = predict_model(tuned_rf, data=data_unseen)
metrics2_2 = pull()  # <-- inmediatamente después
```

```python
print(metrics1_2)
print(metrics2_2)
```
